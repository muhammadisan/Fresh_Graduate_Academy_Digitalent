```
from sklearn.cluster import KMeans

inertia = []
for k in range(2,11):
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(data[['Annual Income (k$)', 'Spending Score (1-100)']])
  inertia.append(kmeans.inertia_)
```

```
fig, ax = plt.subplots(figsize=(9,6))

ax.plot(range(2,11), inertia)
ax.set_title('Elbow Method Kmeans', fontsize=17)
ax.set_ylabel('inertia')
ax.set_xlabel('k in k-means')

plt.show()
```

```
kmeans = KMeans(n_clusters=5)
kmeans.fit(data[['Annual Income (k$)', 'Spending Score (1-100)']])
```

```
kmeans.cluster_centers_
```

```
data['label'] = kmeans.labels_
```

```
fig, ax = plt.subplots(figsize=(7,7))

ax.scatter(data[data['label'] == 0]['Annual Income (k$)'], data[data['label'] == 0]['Spending Score (1-100)'],
           color='red', label = 'YOLO')
ax.scatter(data[data['label'] == 1]['Annual Income (k$)'], data[data['label'] == 1]['Spending Score (1-100)'],
           color='blue', label = 'Sultan (Target)')
ax.scatter(data[data['label'] == 2]['Annual Income (k$)'], data[data['label'] == 2]['Spending Score (1-100)'],
           color='yellow', label = 'Orang Biasa')
ax.scatter(data[data['label'] == 3]['Annual Income (k$)'], data[data['label'] == 3]['Spending Score (1-100)'],
           color='green', label = 'Pelit')
ax.scatter(data[data['label'] == 4]['Annual Income (k$)'], data[data['label'] == 4]['Spending Score (1-100)'],
           color='cyan', label = 'Hemat')

ax.set_xlabel('Annual Income (k$)')
ax.set_ylabel('Spending Score')
ax.set_title('Cluster by Annual Income and Spending Score', fontsize=17)
ax.legend()
plt.show()
```
