```
import pandas as pd

df2017q1 = pd.read_csv('drive/My Drive/FGA/Toronto Bikeshare Data/2017 Data/Bikeshare Ridership (2017 Q1).csv')
df2017q2 = pd.read_csv('drive/My Drive/FGA/Toronto Bikeshare Data/2017 Data/Bikeshare Ridership (2017 Q2).csv')
df2017q3 = pd.read_csv('drive/My Drive/FGA/Toronto Bikeshare Data/2017 Data/Bikeshare Ridership (2017 Q3).csv')
df2017q4 = pd.read_csv('drive/My Drive/FGA/Toronto Bikeshare Data/2017 Data/Bikeshare Ridership (2017 Q4).csv')
df2017 = pd.concat([df2017q1, df2017q2, df2017q3, df2017q4]).reset_index().drop('index', axis=1)
df2017
```

**trip_duration ada yang sampai detik dan ada yang tidak. Tridak ada station_id di Q3 dan Q4**

```
df2018q1 = pd.read_csv('drive/My Drive/FGA/Toronto Bikeshare Data/bikeshare2018/Bike Share Toronto Ridership_Q1 2018.csv')
df2018q2 = pd.read_csv('drive/My Drive/FGA/Toronto Bikeshare Data/bikeshare2018/Bike Share Toronto Ridership_Q2 2018.csv')
df2018q3 = pd.read_csv('drive/My Drive/FGA/Toronto Bikeshare Data/bikeshare2018/Bike Share Toronto Ridership_Q3 2018.csv')
df2018q4 = pd.read_csv('drive/My Drive/FGA/Toronto Bikeshare Data/bikeshare2018/Bike Share Toronto Ridership_Q4 2018.csv')
df2018 = pd.concat([df2018q1, df2018q2, df2018q3, df2018q4]).reset_index().drop('index', axis=1)
df2018
```

**Tidak ada null values pada data 2018. Kolom data juga sudah konsisten memasukan station_id, tidak seperti data tahun 2017**

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(1,2, figsize=(20,5))

sns.distplot(df2017[df2017['trip_duration_seconds'] < 3600]['trip_duration_seconds'], ax=ax[0])
sns.distplot(df2018[df2018['trip_duration_seconds'] < 3600]['trip_duration_seconds'], ax=ax[1])

plt.show()

print('minimum trip durations in 2017 is', df2017['trip_duration_seconds'].min(), 'seconds')
print('maximum trip durations in 2017 is', df2017['trip_duration_seconds'].max(), 'seconds')

print('minimum trip durations in 2018 is', df2018['trip_duration_seconds'].min(), 'seconds')
print('maximum trip durations in 2018 is', df2018['trip_duration_seconds'].max(), 'seconds')
```

**Dari distribution plot terlihat cukup banyak trip_duration tidak ada 1 menit di tahun 2017. Dan menarik bahwa ada yang sampai ~6juta seconds atau sekitar 73 hari. Mari kita lihat lebih lanjut**

```
from datetime import datetime

df2017q1['trip_start_time'] = df2017q1['trip_start_time'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y %H:%M'))
df2017q1['trip_stop_time'] = df2017q1['trip_stop_time'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y %H:%M'))
```

```
df2017q1.head()
```

```
df2017q1['start_day'] = df2017q1['trip_start_time'].apply(lambda x: datetime(x.year, x.month, x.day))
df2017q1['stop_day'] = df2017q1['trip_start_time'].apply(lambda x: datetime(x.year, x.month, x.day))
```

```
df = df2017q1.groupby('start_day').count()
```

```
fig, ax = plt.subplots(figsize=(20,5))
plt.bar(x=df.index, height=df['trip_id'])
plt.show()
```
